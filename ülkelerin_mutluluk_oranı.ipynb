```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight") # fivethirtyeight data science da kullanılan tipik bir görselleştirme

import warnings
warnings.filterwarnings("ignore")  #Bu komut, belirli uyarıların konsol veya terminal çıktısında görünmesini engeller

from sklearn.linear_model import LinearRegression
```

```python
#dataseti kaggle'dan aldım, kaggle daki datasetinde 2019 dan sonraki yılların raporlarını benzer şekilde
# bulamadığım için 2019 dan sonraki yıllar yok.

df1 = pd.read_csv("2015.csv")     
df2 = pd.read_csv("2016.csv")
df3 = pd.read_csv("2017.csv")
df4 = pd.read_csv("2018.csv")
df5 = pd.read_csv("2019.csv")


print(df1.columns,"\n",df2.columns,"\n",df3.columns,"\n",df4.columns,"\n",df5.columns)
```

```python
df1.drop(["Region"], axis=1, inplace=True)
df1["Year"] = 2015
# aşağıdaki sütun isimleri anlamlı ve analize daha uygun olarak değiştirildi
df1.rename(columns={"Happiness Rank": "Rank",
            "Happiness Score": "Score", 
            "Standard Error": "Error", 
            "Family": "social_support",
            "Economy (GDP per Capita)": "Economy", 
            "Health (Life Expectancy)": "life_expectancy", 
            "Trust (Government Corruption)": "Corruption", 
            "Dystopia Residual": "dsytopia_residual"}, 
            inplace=True)

  # aşağıda Error sütunu, Upper Confidence Interval ve Lower Confidence Interval sütunlarının farkı alınarak hesaplandı ve eklendi
df2["Error"] = df2["Upper Confidence Interval"] - df2["Lower Confidence Interval"]
# aşağıda Region, Upper Confidence Interval, ve Lower Confidence Interval sütunları kaldırılır.
df2.drop(["Region","Upper Confidence Interval","Lower Confidence Interval"], axis=1, inplace=True)
df2["Year"] = 2016
# aşağıdaki sütun isimleri df 1 deki gibi düzenlendi
df2.rename(columns={"Happiness Rank": "Rank", 
            "Happiness Score": "Score", 
            "Family": "social_support",
            "Economy (GDP per Capita)": "Economy", 
            "Health (Life Expectancy)": "life_expectancy", 
            "Trust (Government Corruption)": "Corruption", 
            "Dystopia Residual": "dsytopia_residual"}, 
            inplace=True)
# aşağıdaki Error sütunu, Whisker.high ve Whisker.low sütunlarının farkı alınarak hesaplanır ve eklenir
df3["Error"] = df3["Whisker.high"] - df3["Whisker.low"]

# aşağıda Whisker.high ve Whisker.low sütunları kaldırılır.
df3.drop(["Whisker.high","Whisker.low"], axis=1, inplace=True)
df3["Year"] = 2017
df3.rename(columns={"Happiness.Rank": "Rank", 
            "Happiness.Score": "Score", 
            "Family": "social_support",
            "Economy..GDP.per.Capita.": "Economy", 
            "Health..Life.Expectancy.": "life_expectancy", 
            "Trust..Government.Corruption.": "Corruption", 
            "Dystopia.Residual": "dsytopia_residual"}, 
            inplace=True)

df4["Year"] = 2018
df4.rename(columns={"Overall rank": "Rank", 
            "Country or region": "Country",
            "GDP per capita": "Economy", 
            "Social support": "social_support",
            "Healthy life expectancy": "life_expectancy", 
            "Freedom to make life choices": "Freedom", 
            "Perceptions of corruption": "Corruption"}, 
            inplace=True)

df5["Year"] = 2019
df5.rename(columns={"Overall rank": "Rank", 
            "Country or region": "Country",
            "GDP per capita": "Economy", 
            "Social support": "social_support",
            "Healthy life expectancy": "life_expectancy", 
            "Freedom to make life choices": "Freedom", 
            "Perceptions of corruption": "Corruption"}, 
            inplace=True)

```

Verileri aynı özniteliğe denk gelen kolon adlarını düzenledim, şimdi istenen özniteliklerine göre birleştirelim

```python
# aşağıda df1den df5e kadar olan beş farklı DataFrame'den seçilen sütunları kullanarak, hepsini tek bir DataFrame'e birleştirir

cols = ["Rank","Country","Economy","social_support","life_expectancy","Freedom","Generosity","Corruption","Year"]
data = pd.concat([df1[cols],df2[cols],df3[cols],df4[cols],df5[cols]], axis=0).reset_index(drop=True)
data.sample(10) # birleştirilmiş veri setinden rastgele 10 örnek
```

```python
data.isnull().sum()
```

```python
#Corruption kolonunda 1 null değerimiz var bunlara bakalım:
data[data["Corruption"].isnull()]
```

```python
# Sadece 1 tane null girdi var ve bunu silmek yerine uygun değerle doldurmak için UAE'ye ait diğer girdilere bakıp doldurmak daha iyi bir seçenektir diye düşünüyorum
data[data["Country"] == "United Arab Emirates"]
```

```python
# Boş değerleri doğrusal regresyon tahminleriyle doldurma

d = data[data["Country"] == "United Arab Emirates"][["Corruption","Year"]]

plt.figure(figsize=(10,7)) # 10 inç genişliğinde ve 7 inç yüksekliğinde
plt.plot(d["Year"], d["Corruption"], "*", color="red") # d dataFramenindeki corrupation değerlerini çizer
plt.axis([2014,2020, 0,0.5]) # eksenlerin sınırlarını ayarlama
plt.title("Corruption - United Arab Emirates") # başlık
plt.show()
```

```python
#altta regresyon ile değer tahmini yapıyorum

X = d[d["Year"] != 2018]["Year"].values.reshape(-1, 1)
y = d[d["Year"] != 2018]["Corruption"].values

reg = LinearRegression().fit(X,y) # model eğitme
print("2018 için tahmin: ", reg.predict(np.array(2018).reshape(-1, 1)))
```

```python
#aşağıda Corruption" sütununda eksik değerlere (NaN) sahip olan satırları seçer.
#.index kullanılarak, bu satırların indeksleri alınır ve i değişkenine atanır.
i = data[data["Corruption"].isnull()].index

#Önceki adımda oluşturulan lineer regresyon modeli (reg) kullanılarak, 2018 yılı için bir tahmin yapılır. 
#Bu tahmin, 2018 yılına karşılık gelen yıllık yolsuzluk oranını temsil eder.
data["Corruption"][i] = reg.predict(np.array(2018).reshape(-1, 1))
data[data["Country"] == "United Arab Emirates"]

#yukarıdaki kod parçaları, eksik yolsuzluk değerlerini tahmin edilmiş değerlerle doldurarak veri setindeki eksiklikleri giderir.
```

```python
print("Toplam null girdi sayısı:", data.isnull().sum().sum())
```

```python
# Ülkelere göre gruplama ve puanlarını sıralıyorum

grouped_country = data.groupby("Country").mean()
```

# En yüksek puana sahip ülkeler


```python
fig = plt.figure(figsize=(15,20))

rows = 3
columns = 2

for i,column in enumerate(grouped_country.columns.values[1:-1]): #gruplanmış veri setindeki sütunların üzerinde döngü oluşturma
    if column != "Corruption":
        sorted_grouped_country = grouped_country.sort_values(column, ascending=False)[:10]
    else:
        sorted_grouped_country = grouped_country.sort_values(column)[:10]
    plt.subplot(rows,columns,i+1)
    sorted_grouped_country[column].plot(kind="barh", color="orange") #yatay çubuk grafiklerini oluşturur
    plt.ylabel("") #y eksenine ait etiket yok
    plt.title(column) 
```

# En düşük puana sahip ülkeler


```python
fig = plt.figure(figsize=(15,20))

rows = 3
columns = 2

for i,column in enumerate(grouped_country.columns.values[1:-1]):
    if column != "Corruption":
        sorted_grouped_country = grouped_country.sort_values(column, ascending=False).tail(10)
    else:
        sorted_grouped_country = grouped_country.sort_values(column).tail(10)
    plt.subplot(rows,columns,i+1)
    sorted_grouped_country[column].plot(kind="barh", color="green")
    plt.ylabel("")
    plt.title(column)
```

# Yıllara göre değişiklikler


```python
grouped_years = data.sort_values(["Year","Rank"]).groupby("Year").head(10).reset_index(drop=True)
grouped_years #her bir yıldan en üst 10 ülkeyi içeren bir veri setini temsil eder.
```

yukarıdaki tabloyu aşağıda görselleştirerek daha az göz yoran ve anlaması kolay olacak hale getirelim

```python
colors = ["Blue","Red","Yellow","Black","Green","Orange","Pink","Purple","Grey","Turquoise"]

fig = plt.figure(figsize=(15,20))

for i,country in enumerate(grouped_years["Country"].unique()[:10]):
    plt.subplot(5,2,i+1) # 5 satır 2 sütun
    #aşağıdaki s=100 parametresi, noktaların boyutunu ayarlıyo.
    plt.scatter(grouped_years[grouped_years["Country"] == country]["Year"], grouped_years[grouped_years["Country"] == country]["Rank"], s=100, marker="*", linestyle="dotted", c=colors[i])
    plt.plot(grouped_years[grouped_years["Country"] == country]["Year"], grouped_years[grouped_years["Country"] == country]["Rank"], linestyle="dotted", c=colors[i])
    plt.axis([2014,2019, 0,11])
    plt.title(country)
    plt.suptitle("Top 10 country rankings through years")
    plt.tight_layout()
```

```python
# tail(10) ile en her bir yılın en alttan 10 ülkesini seciyoruz
grouped_years_ = data.sort_values(["Year","Rank"]).groupby("Year").tail(10).reset_index(drop=True)
grouped_years_
```

```python
fig = plt.figure(figsize=(15,20))

for i,country in zip(range(10),grouped_years_["Country"].unique()):
    plt.subplot(5,2,i+1)
    plt.scatter(grouped_years_[grouped_years_["Country"] == country]["Year"], grouped_years_[grouped_years_["Country"] == country]["Rank"], s=100, marker="*", linestyle="dotted", c=colors[i])
    plt.plot(grouped_years_[grouped_years_["Country"] == country]["Year"], grouped_years_[grouped_years_["Country"] == country]["Rank"], linestyle="dotted", c=colors[i])
    plt.axis([2014,2019, 145,160])
    plt.title(country)
    plt.suptitle("Country rankings having lowest scores through years")
    plt.tight_layout()
```

Rapora göre en mutlu olan 10 ülke sıralamarında belirgin bir değişim görülmezken, listede son sıralarda yer alan ülkelerde iniş ve çıkışlar ve büyüklükleri daha geniş aralıkta gerçekleşmekte.

En mutlu olarak raporlanmış ülkelerdeki refah durumu, eşitlik ve adalet anlayışı gibi konularda daha stabil bir ortam olması bu ülkelerin konumunu korumasında rol oynamış olabilir.

Son sıralardaki ülkelerde ise özgürlük ve yozlaşmışlık düzeyi gibi faktörlerin yanı sıra verisetinde yer almayan başka etkenler bu sıralamayı etkilemiş olabilir.
